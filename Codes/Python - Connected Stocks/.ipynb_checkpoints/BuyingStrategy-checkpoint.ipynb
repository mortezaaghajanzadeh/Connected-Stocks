```python
def foo(x):
    if len(x)<2:
        return '0'+x
    else:
        return x
```

```python
df['year_of_year'] = df['year_of_year'].astype(str)
df['week_of_year'] = df['week_of_year'].astype(str)
df['week_of_year'] = df['week_of_year'].apply(foo)
df['Year_Month_week']  = df['year_of_year'] + df['week_of_year']
df['year_of_year'] = df['year_of_year'].astype(int)
df['week_of_year'] = df['week_of_year'].astype(int)
df['Year_Month_week'] = df['Year_Month_week'].astype(int)
```

```python
df1 = df[['date', 'jalaliDate', 'id_x', 'id_y','Year_Month',
       'year_of_year','FCA*','Ret_x', 'Ret_y','id']]
merge = df1.groupby(['id','Year_Month'])[['Ret_x', 'Ret_y']].sum().reset_index()
df1 = df1.drop(columns = ['Ret_x', 'Ret_y']).merge(merge,on =['id','Year_Month'])
df1 = df1.drop_duplicates(subset = ['id','Year_Month'] , keep = 'last')
mean = df1.groupby(['Year_Month'])['FCA*'].mean().to_frame().reset_index().rename(columns = {'FCA*':"Weight"})
df1 = df1.sort_values(by = ['Year_Month','FCA*']).merge(mean,on = 'Year_Month')
df1['Weight'] = df1['FCA*'] - df1['Weight'] + 2
df1['Weight'] = df1['Weight']/2
```

```python
df1['Sign_x'] = -1
df1.loc[df1.Ret_x>0,'Sign_x'] = 1
df1['Sign_y'] = -1
df1.loc[df1.Ret_y>0,'Sign_y'] = 1
```

```python
df1.head()
```

```python
def sign(g):
    d = {}
    for i in set(g.id_x)|set(g.id_y):
        t = g.loc[(g.id_x == i)]
        t1 = g.loc[(g.id_y == i)]
        d[i] = (t['Sign_y'] * t['Weight']).sum()+(t1['Sign_x'] * t1['Weight']).sum()
    return d
```

```python
data = {}
gg = df1.groupby('Year_Month')
for i in gg.groups.keys():
    g = gg.get_group(i)
    print(i)
    data[i] = sign(g)
```

```python
dd = pd.DataFrame.from_dict(data).T.to_dict()
```

```python
Results = (df1[['id_x',
     'Year_Month',
     'Ret_x']].rename(columns ={'id_x':'id',
                               'Ret_x':'Ret'})).append(df1[['id_y',
                                                            'Year_Month',
                                                            'Ret_y']].rename(columns ={'id_y':'id',
                                                       'Ret_y':'Ret'})).sort_values(by = ['id','Ret']).drop_duplicates()
Results = Results.groupby(['id','Year_Month'])['Ret'].mean().to_frame().reset_index()
```

```python
FResult = pd.DataFrame()
for i in dd.keys():
    print(i)
    t = pd.Series(dd[i]).to_frame().reset_index().rename(columns = {0:'Pair','index':'Year_Month'})
    t['id'] = i
    FResult = FResult.append(Results.merge(t,on = ['id','Year_Month']))
```

```python
FResult
```

```python
FResult['FRet'] = FResult.groupby('id').Ret.shift(-1)
FResult['Low'] = 0
FResult.loc[(FResult.Ret<0)&(FResult.Pair<0),'Low'] = 1
FResult['High'] = 0
FResult.loc[(FResult.Ret>0)&(FResult.Pair>0),'High'] = 1
```

```python
FResult['LowPRet'] = FResult['FRet'] * FResult['Low']
FResult['HighPRet'] = FResult['FRet'] * FResult['High']
FResult['FPRet'] =  FResult['LowPRet'] - FResult['HighPRet']
```
